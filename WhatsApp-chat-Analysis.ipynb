```python
f = open("WhatsApp Chat with IIIT BBSR'26.txt", 'r', encoding='utf-8') # opening the file
```

```python
# Creating a new data frame
# creating 2 column, date and time in one column and name and msg in another column
# for this use regx101.com
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APMapm]{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
 dates = re.findall(pattern, data)
 dates
# Extracting dates 
```

```python
import pandas as pd

# Assuming 'messages' and 'dates' are already defined
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M\u202f%p - ')

# Rename the 'message_date' column to 'date'
df.rename(columns={'message_date': 'date'}, inplace=True)

# Display the first few rows of the DataFrame
df.head()

```

```python
df.shape  # how many messages- 13617
```

```python
    # User name and msg are in the same column so we changed it
    users = []
    messages = []
    for message in df['user_message']:
        entry = re.split('([\w\W]+?):\s', message)
        if entry[1:]:  # user name
            users.append(entry[1])
            messages.append(" ".join(entry[2:]))
        else:
            users.append('group_notification')
            messages.append(entry[0])

    df['user'] = users
    df['message'] = messages
    df.drop(columns=['user_message'], inplace=True)
    df.head()
```

```python
    #separating date, year,month,dayname,hr and minute
    df['only_date'] = df['date'].dt.date
    df['year'] = df['date'].dt.year
    df['month_num'] = df['date'].dt.month
    df['month'] = df['date'].dt.month_name()
    df['day'] = df['date'].dt.day
    df['day_name'] = df['date'].dt.day_name()
    df['hour'] = df['date'].dt.hour
    df['minute'] = df['date'].dt.minute
    df.head()
```

```python
period= []
for hour in df[['day_name', 'hour']][ 'hour' ]:
    if hour == 23:
        period.append(str(hour)+ "-" +str('00'))
    elif hour == 0:
        period.append(str ('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
